# # Preprocessing the Data

```python
df.isnull().sum()   #Number of NULL data points
```

```python
df.drop('Married',axis=1)
df.drop('Education',axis=1)
df.drop('Loan_ID',axis=1)
```

```python
df['Self_Employed'] = df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
df['LoanAmount'] = df['LoanAmount'].fillna(df['LoanAmount'].median())
df['Loan_Amount_Term'] = df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].median())
df['Credit_History'] = df['Credit_History'].fillna(df['Credit_History'].median())
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df['Dependents'] = df['Dependents'].fillna(df['Dependents'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.loc[df["Gender"] == "Male", "Gender"] = 0
df.loc[df["Gender"] == "Female", "Gender"] = 1
```

```python
df.loc[df["Education"] == "Not Graduate", "Education"] = 0
df.loc[df["Education"] == "Graduate", "Education"] = 1
```

```python
df.loc[df["Loan_Status"] == "N", "Loan_Status"] = 0
df.loc[df["Loan_Status"] == "Y", "Loan_Status"] = 1
```

```python
df.loc[df["Self_Employed"] == "No", "Self_Employed"] = 0
df.loc[df["Self_Employed"] == "Yes", "Self_Employed"] = 1
```

```python
from sklearn.preprocessing import LabelEncoder

depen = ['Dependents']
le = LabelEncoder()
for i in depen:
    df[i] = le.fit_transform(df[i])
    
pro_area = ['Property_Area']
le = LabelEncoder()
for i in pro_area:
    df[i] = le.fit_transform(df[i])

```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop(['Loan_Status','Loan_ID','Married'],axis=1)
y=df['Loan_Status']
```

## Splitting the Data for Training and Testing :-

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=100)    #Splitting the data
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## USING LOGISTIC REGRESSION :-

```python
 from sklearn.linear_model import LogisticRegression
```

```python
#Training
my_model = LogisticRegression()

results = my_model.fit(X_train,y_train)
```

```python
#Testing
predictions = results.predict(X_test)
predictions
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy using Logistic Regeression is : ',accuracy_score(y_test, predictions))   #Accuracy of Model
```

```python
from sklearn import metrics
```

```python
print('\n**Classification Report:\n',metrics.classification_report(y_test,predictions))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_mat = confusion_matrix(y_test, predictions)
```

```python
confusion_df = pd.DataFrame(confusion_mat, index=['Actual 0','Actual 1'], columns=['Predicted 0','Predicted 1'])
```

```python
sns.heatmap(confusion_df, cmap='inferno_r', annot=True)
```
